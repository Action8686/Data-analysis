# Matplotlib 的figure基本使用

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 如果不创建figure对象，matplotlib将会自动的创建一个figure对象
fig = plt.figure()
print(type(fig))
```

```python
arr = np.random.randn(100)
```

```python
# 可以通过add_subplot来分隔figure,表示可以在figure的不同位置上作图
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2 ,3)

ax3 = fig.add_subplot(2, 2 ,4)
```

```python
ax2.plot(arr)
ax3.plot(arr)
ax1.plot(arr)
plt.show()
```

# 直方图hist

```python
arr = np.random.randn(100)
# 第一个参数是数据集，hist参数代表展现数据的直方个数，color参数可以指定颜色，alpha参数可以指定透明度（默认是1，表示不都透明）
plt.hist(arr, bins = 20, color='c', alpha=0.5)
plt.show()
```

## 散点图 scatter 


```python
# x表示x轴的标签
x  =[1,2,3,4]
# y表示y轴的标签
y = [2,4,6,8]
#创建散点图
plt.scatter(x, y)
plt.show()
```

```python
# x 表示100个 横坐标，100个数据
x=np.arange(100)
# y 表示纵坐标， 100个数据，但是经过运算
y=x + 2 * np.random.randn(100)
plt.scatter(x,y)
plt.show()
```

## 柱形图 bar

```python
x = np.arange(10)
y1, y2 = np.random.randint(10, 50, size=(2, 10))               
```

```python
# 创建一个 8 * 5大小的图，指定每个点的分辨率是100 （默认是80）
# 最终图像大小是： 8*100 * 5*100 = 800 * 500
fig = plt.figure(figsize = (8,5), dpi=100)
ax=plt.subplot(1,1,1)
```

```python
# 宽度偏移量
width = 0.25

# x 表示柱子所在的横坐标位置， y1 表示纵坐标位置，width表示柱子的宽度(默认是0.8 英寸),color 指定黄色
ax.bar(x, y1, width, color = 'y', alpha = 0.5)
# x + width 表示基于原先位置增加 width偏移量（刚好是y1柱子的宽度), y2 表示纵坐标位置，color 指定红色
ax.bar(x + width, y2, width, color='r', alpha = 0.5)

# 指定 x轴标记的位置
ax.set_xticks(x + width / 2)
ax.set_xticklabels(['a','b','c','d','e','f','g','h','i','j'])
plt.savefig("bar.png")
plt.show()
```

## 刻度、标签、图例

```python
# 返回一个fig 和 axes 对象(唯一一个)
fig, ax = plt.subplots(1)
```

```python
# 设置刻度范围
ax.set_ylim([0, 1000])

# 设置x轴显示的刻度值
ax.set_yticks(range(0, 1000, 200))

# 设置y轴显示的刻度值，十二个月的英文简写
ax.set_xticklabels(['Jan', 'Feb','Mar','Apr','May','Jun','Jul','Aug', 'Sep', 'Oct', 'Nov','Dec'])
```

```python
# 设置 x轴的标签
ax.set_ylabel("Money")
# 设置 y轴的标签
ax.set_xlabel("Month")
```

```python
# 设置整个绘图的title （年度报告）
ax.set_title("Annual Statement")
```

```python
ax.plot(np.random.randint(300, 880, 12), label = u"huangmenji", color='r')
ax.plot(np.random.randint(300, 800, 12), label = u"kaoroubanfan")
ax.plot(np.random.randint(300, 800, 12), label = u"huangqiaoshaobing")
```

```python
# 显示图例
ax.legend()
plt.savefig("eat.png")
#显示绘图结果
plt.show()
```
